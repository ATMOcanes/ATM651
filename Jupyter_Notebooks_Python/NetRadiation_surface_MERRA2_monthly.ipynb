```python
# xarray install may have some options, I recommend checking it and installing with its dependencies

import xarray as xr
import matplotlib.pyplot as plt
```

```python
ds = xr.open_dataset('https://goldsmr4.gesdisc.eosdis.nasa.gov/dods/M2TMNXRAD')
```

```python
ds
```

```python
netrad1980 = ds.lwgnt[0:12]+ds.swgnt[0:12]
netrad2020 = ds.lwgnt[12*40:12*40+12]+ds.swgnt[12*40:12*40+12]
```

```python
plt.figure(figsize=[20,15])

for i in range(12):
    plt.subplot(4,3,i+1)
    netrad1980[i].plot(clim=[-200,200])
    netrad1980[i].plot.contour(levels=[0])
    netrad2020[i].plot.contour(levels=[0],colors='g')
    if i>0: 
        plt.axis('off')
    # plt.ylim([0,70]); plt.xlim([-150,-10]) # North Amer-atlantic
    plt.ylim([-70,70]) # world North Amer-atlantic
```
