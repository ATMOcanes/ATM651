```python
files = glob('/Users/brianmapes/Box/RCEMIP/1Dprofiles/*300*.nc')
for fi in files: 
    ds = xr.open_dataset(fi)
    plt.plot(ds.ta_avg, ds.pa_avg)
plt.ylim([1000,100]); plt.xlim([200,300])
plt.title('300K small RCE mean T profiles')
```

```python
files = glob('/Users/brianmapes/Box/RCEMIP/1Dprofiles/*300*.nc')
for fi in files: 
    ds = xr.open_dataset(fi)
    plt.plot(ds.ta_avg-ds.ta_avg.sel(zg_avg = 0, method='nearest'), ds.pa_avg)
plt.ylim([1000,100]); plt.xlim([-100,0])
plt.title('300K small RCE mean T profiles minus z=0 value')
```

```python
files = glob('/Users/brianmapes/Box/RCEMIP/1Dprofiles/*300*.nc')
for fi in files: 
    ds = xr.open_dataset(fi)
    plt.scatter(ds.ta_avg.sel(zg_avg = 0, method='nearest'), 
                ds.ta_avg.sel(zg_avg = 5, method='nearest') )
# plt.ylim([1000,100])
plt.title('300K small RCE T0m vs. T5km')
```

```python
files = glob('/Users/brianmapes/Box/RCEMIP/1Dprofiles/*300*.nc')
for fi in files: 
    # print(fi)
    ds = xr.open_dataset(fi)
    if('hus_avg' in list(ds.keys())): 
        plt.scatter(ds.ta_avg.sel(zg_avg = 0, method='nearest'), 
                    ds.hus_avg.sel(zg_avg = 0, method='nearest') )
plt.ylim([10,18])
plt.title('300K small RCE T0m vs. q0m')
```
