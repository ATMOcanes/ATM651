```python
import pandas as pd
import datetime
import csv
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
df = pd.read_csv('./ConsVariables_nearGhats_July_41y.csv') # ,parse_dates=     {"Datetime" : [1,2]},names=headers)
df
```

```python
years = df.Time.values
years
```

```python
year = [int(st[0:4]) for st in years]
```

```python
plt.figure(figsize=(15, 10))

plt.scatter(df.thetaesat, -df.Level, c=year, cmap='jet', marker='o')
plt.xlim([325,370])
plt.title('Years from 1980-2021, Theta-e-sat is temperature')
plt.colorbar()
#plt.plot(df.thetaesat, -df.Level, c=year)
```

```python
plt.figure(figsize=(15, 10))

plt.scatter(df.thetae, -df.Level, c=year, cmap='jet', marker='^', alpha=0.2)
plt.xlim([325,370])
plt.title('Years from 1980-2021, Theta-e is temperature')
plt.colorbar()
#plt.plot(df.thetaesat, -df.Level, c=year)
```
