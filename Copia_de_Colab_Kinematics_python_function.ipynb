### Background
**Kinematics** is not dynamics, it is just a *description* of how 2D vector fields $\vec V_{2D}$ = **V** = \<u,v\> vary in a 2D space with Cartesian coordinates x,y. 

* Consider all possible LINEAR dependences of **V** = \<u,v\> on space (x,y) around an origin (0,0)
    * Measured by 4 coefficients $c_1, c_2, c_3, c_4$
* Also, add a constant (that is, not varying with x,y) vector field U,V. 

In other words, our *displayed* vector field V(x,y) = {u,v}(x,y) is: 

$$ u_{disp} = U + c_1 x  +c_2 y$$
$$ v_{disp} = V + c_3 x  +c_4 y$$

## Exercise 2: Create and display the basic fields in Python
* The first code cell (shift-return to execute it and move on) sets up a coordinate system (x,y) from -10 to 10.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set up a coordinate grid centered on 0,0
n = 10
x,y = np.mgrid[-n:n+1, -n:n+1]
```

* Now define the basic set of "pure" div, vort, def1, def2, and a mean flow in the x or y directions (U, V). 
* Note that each field has a u and a v component indicated by the first letter
  * like udiv and vvor and uV=0 and vU=0
* Notice that both components are set on one line using the comma **,**  
* Study these names and understand their logic, You will need it, in order to construct sums of these in the cells below. 
* Finally, execute the cell that displays the vector fields as arrows using the *quiver* function. 
* The first one has been done for you, then you complete the work for the other cases by example.  

```python
# Pure divergence of unit magnitude
udiv,vdiv = 1*x, 1*y

# Pure vorticity of unit magnitude
uvor,vvor = -1*y, 1*x

# Pure def1 of unit magnitude
udef1,vdef1 = 1*x, -1*y

# Pure def2 of unit magnitude
udef2,vdef2 = 1*y, 1*x

# Pure U and V
uU, vU = 1, 0
uV, vV = 0, 1
```

```python
# a helper function to make the standard plot 
def plotquiver(udisp,vdisp,label):
  fig, ax = plt.subplots(figsize=(7,7))
  plt.quiver(x, y, udisp, vdisp, pivot='mid')
  plt.title(label)
  plt.show()
```

# First combination: pure divergence, plus a little mean U wind

```python
# Define the display field udisp,vdisp and give it a label. 

udisp,vdisp = udiv + 2*uU, vdiv + 2*vU
label = 'pure div strength 1 plus U=2'
plotquiver(udisp,vdisp,label)
```

#Pure vorticity

```python
# Define the display field udisp,vdisp and give it a label. 

udisp,vdisp = uvor,vvor
label = 'Pure unit vorticity'
plotquiver(udisp,vdisp,label)
```

# Pure deformation type 1 

```python
# Define the display field udisp,vdisp and give it a label. 

udisp,vdisp = udef1,vdef1 
label = 'Pure unit def1'
plotquiver(udisp,vdisp,label)
```

# Pure deformation type 2


```python
# Define the display field udisp,vdisp and give it a label. 

udisp,vdisp = udef2,vdef2 
label = 'Pure unit def2'
plotquiver(udisp,vdisp,label)
```

# Sum of def1 + def2 (equal parts, or 2 to 1, or 1 to 2). Experiment! 


```python
# Define the display field udisp,vdisp and give it a label. 

udisp,vdisp = udef2+udef1, vdef2+vdef1
label = '1*def2 + 1*def1'
plotquiver(udisp,vdisp,label)
```

# Sum of vorticity plus deformation in equal measure

Note that these are two curved flow patterns. Together they makes a flow we could describe as... (YOUR ANSWER HERE after you find out)
